```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
from pyodide.http import pyfetch

async def download (url, filename):
    response=await pyfetch(url)
    if response.status==200:
        with open(filename, 'wb') as f:
            f.write(await response.bytes())
    
```

```python
file_path="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
await download(file_path,'usedcars.csv')
file_name ='usedcars.csv'
```

```python
df=pd.read_csv(file_name, header=None)
df
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns=headers
```

```python
df
```

```python
import numpy as np
df.replace('?',np.NaN, inplace=True)
df.head()
```

```python
avg_norm_loss=df["normalized-losses"].astype('float').mean(axis=0)
print('Average of normalized-losses:',avg_norm_loss)
```

```python
import pandas as pd
df["normalized-losses"].replace(np.nan,avg_norm_loss,inplace=True)
df
```

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
print('Average of bore:',avg_bore)
```

```python
df['bore'].replace (np.nan,avg_bore,inplace=True)
df
```

```python
Avg_stroke=df['stroke'].astype('float').mean(axis=0)
print('Average stroke:',Avg_stroke)
```

```python
df['stroke'].replace(np.nan,Avg_stroke)
df
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#df['horsepower']
#df['horsepower']=df['horsepower'].astype('int').mean()
df['horsepower'] = df['horsepower'].fillna(0).astype(int)
df['horsepower']
```

```python
bins=np.linspace(min(df['horsepower']),max(df['horsepower']),4)
bins
```

```python
group_names= ['Low','Medium','High']
```

```python
df['horse-binned']=pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df['horse-binned'].head(20)
```

```python
df['horse-binned'].value_counts()
```

```python
df['horse-binned'].value_counts().idxmax()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df['horse-binned'].value_counts())
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
